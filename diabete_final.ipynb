#### Graphiques

Puisque nous allons réutiliser plusieurs fois certains types graphiques, nous les plaçons dans des fonctions : 

````r
# Graphiques en barres horizontal montrant les proportions de la variables va pour 
# en fonction du statut de diabète de la personne 
graph_horizontal_bars_diabete <- function(df, diabete, var) {
  
    # Define colors as a constant vector
    colors <- c('#ff5809', '#1E88E5')
  
    # Use `table()` to calculate contingency table
    table <- table(df[[diabete]], df[[var]])
  
    # Sum the rows of the contingency table
    sum_table <- rowSums(table)
  
    # Calculate the percentage table by dividing each cell by its row sum
    var_count <- table/sum_table
  
    # Reverse the order of the columns in the percentage table
    var_count <- var_count[, ncol(var_count):1]
  
    # Printing the name of the dataframe
    #print(colnames(df)[1])
    
    # Making some arangements
    if (colnames(df)[1] == "Diabetes_012") {
        names <- c("Non-Diabetic", "Pre-Diabetic", "Diabetic")
    }
    else {
        names <- c("Non-Diabetic", "Diabetic")
    }
    
    # Use `barplot()` to create a stacked bar chart
    # Add descriptive labels to the chart
    # Use `barplot()` to create a stacked bar chart
    # Add descriptive labels and percentage values to the chart
    bp <- barplot(t(as.matrix(var_count)), # Transpose for having the column sum equal to 1
        horiz = TRUE, 
        col = colors, 
        xlim = c(0,1), 
        xlab = "Percentage", 
        ylab = "", 
        main = paste("Percentage of", var, "by",diabete,"status"), 
        names.arg = names, 
        legend = TRUE,
        args.legend = list(title="Réponses", box.col = "black", bg = "white", legend=c("Oui", "Non")))
    
    #print(var_count)
    # Add percentage values to the right side of the chart using the `text()` function
    par(mar = c(5, 7, 4, 2) + 0.1) # Increase the right margin to create more space for the labels
    text(x = var_count[,1] , y = bp, labels = paste0(round(var_count[,1]*100),"%"), col = "white", pos = 2)
}


# Fonction qui affiche des camembert :
graph_pie <- function(data, variable) {
  # Calculer les fréquences de chaque valeur de la variable
  freq <- table(data[[variable]])
  
  # Calculer les pourcentages correspondants
  percent <- round(prop.table(freq) * 100, 1)
  
  # Tracer le camembert avec les pourcentages
  pie(freq, 
    main = paste("Fréquences de", variable), 
    labels = paste(names(freq), " (", percent, "%)", sep = ""),
    col = rainbow(length(freq)))
}

# Fonction qui affiche la correlation map du dataframe passé en paramètre :
correlation_map <- function(df) {
    # Set up the data and colormap
    data_heatmap <- df
    custom_cmap <- colorRampPalette(c('#011b7a', 'white', '#850505'))(100)

    # Create a correlation matrix
    corr_matrix <- cor(data_heatmap)

    # Create a mask to hide the upper triangle of the matrix
    mask <- upper.tri(corr_matrix)

    # Melt the matrix into a data frame
    corr_df <- melt(corr_matrix)

    # Create the heatmap plot
    ggplot(corr_df, aes(x = Var1, y = Var2, fill = value)) +
      geom_tile(aes(width = 1, height = 1), color = "white", alpha = 0.8) +
      scale_fill_gradientn(colors = custom_cmap, limits = c(-0.8, 0.8)) +
      theme_minimal() +
      theme(axis.text.x = element_text(angle = 45, hjust = 1, size = 15),
            axis.text.y = element_text(size = 15),
            axis.title = element_blank(),
            legend.position = "none") +
      labs(title = "Correlation Heatmap", subtitle = "", x = "", y = "") +
      coord_fixed() +
      guides(fill = guide_colorbar(barwidth = 25, barheight = 0.5, 
                                   title.position = "top", title.vjust = 0.5, 
                                   label.theme = element_text(size = 15)))

}
````

*Rappel*

$$ logit(P(X)) = ln(\frac{P(X)}{1-P(X)}) = \beta_0 + \beta_1X_1 + \beta_2X_2 + ... + \beta_nX_n $$

- Call: La formule de la régression logistique qui a été utilisée.

- Deviance Residuals: Un résumé des résidus de la régression, qui est une mesure de l'ajustement global du modèle. Les valeurs plus élevées de résidus indiquent que le modèle n'a pas bien prédit les données.

- Coefficients: Les coefficients de régression estimés pour chaque variable explicative (ou prédicteur) incluses dans la régression. Pour chaque variable, l'estimation de la régression est donnée, ainsi que son écart-type, un test de Wald (z-value) et une p-value. Les variables ayant une p-value inférieure à un seuil (habituellement 0.05) sont considérées comme significatives. 
    $$ estimate_0 = intercept = \beta_0 $$
    $$ estimate_i = \beta_i $$
    - $Std. Error$ : l'erreur standard de l'estimation des coeffs
    - $z value$ : la statistique de test z, qui mesure la déviation de l'estimation du coefficient de régression par rapport à sa valeur nulle (zéro) en termes d'unités d'écart-type
    - $Pr(>|z|)$ : la valeur p associée à la statistique de test z, qui indique la probabilité d'obtenir une statistique de test aussi extrême que la valeur observée, sous l'hypothèse nulle que le coefficient de régression est égal à zéro. Si cette probabilité est inférieure à un seuil de significativité choisi (par exemple, 0,05), on peut rejeter l'hypothèse nulle et conclure que le coefficient de régression est significativement différent de zéro.

- Signif. codes: Les codes des significativités des p-values calculées. "étoile étoile étoile" signifie que la p-value est inférieure à 0,001, ", ," que la p-value est inférieure à 0,01, et "," que la p-value est inférieure à 0,05. Les points "." indiquent que la p-value est inférieure à 0,1, tandis que les espaces indiquent que la p-value est supérieure à 0,1. (regarder l'output, c'est chiant les * en markdown)

- Dispersion parameter: Le paramètre de dispersion de la famille de distribution binomiale utilisée pour ajuster le modèle.

- Null deviance: La deviance du modèle de régression nul, c'est-à-dire un modèle qui ne prend en compte aucune variable explicative (les $X_i$)

- Residual deviance: La deviance du modèle final, qui tient compte de toutes les variables explicatives incluses dans le modèle.

- AIC: Le critère d'information d'Akaike (AIC) qui est une mesure de l'ajustement du modèle aux données. Les modèles ayant des valeurs AIC plus faibles sont considérés comme meilleurs.

- Number of Fisher Scoring iterations: Le nombre d'itérations nécessaires pour estimer les coefficients de régression.

````r
# Get the coefficient estimates and their standard errors
coefficients <- summary(logistic_regression)$coef[,1]
std_errors <- summary(logistic_regression)$coef[,2]

# Create a data frame with the coefficients and standard errors
coeff_1 <- data.frame(coefficients, std_errors)

# Get the indices of the three largest coefficients
top_3 <- order(coefficients, decreasing = TRUE)[1:3]

# Get the indices of the three smallest coefficients
bottom_3 <- order(coefficients, decreasing = FALSE)[1:3]

# Create a vector to hold the colors for each bar
bar_colors <- rep("blue", nrow(coeff_1))
bar_colors[top_3] <- "red"
bar_colors[bottom_3] <- "purple"


# Plot the coefficients with error bars
ggplot(coeff_1, aes(x = rownames(coeff_1), y = coefficients, fill = bar_colors)) +
  geom_bar(stat = "identity", alpha = 0.5) +
  geom_errorbar(aes(ymin = coefficients - 2 * std_errors, ymax = coefficients + 2 * std_errors), width = 0.4) +
  scale_fill_manual(values = c("blue","purple", "red")) +
  theme_bw() +
  labs(x = "Predictor Variables", y = "Coefficient Estimate", title = "Logistic Regression Coefficient Estimates") +
  coord_flip()
````

$$ \text{Odds ratio}_i = OR_i = \frac{\text{Odds for disease with risk present}}{\text{Odds for disease with risk absent}} = e^{\beta_i} $$

$$ OR_i = \frac{\frac{P(Y=1|X=1)}{1 - P(Y=1|X=1)}}{\frac{P(Y=1|X=0)}{1 - P(Y=1|X=0)}}$$

- $OR_i = 1$ : le succès est indépendant de la présence de la variable $X_i$
- $OR_i > 1$ : le succès est plus probable avec la présence de la variable $X_i$
- $OR_i < 1$ : le succès est moins probable avec la présence de la variable $X_i$

Succès en question : Devenir le roi des diabètiques 

````r
# Get the odds ratios and their 95% confidence intervals
odds_ratios <- exp(coef(logistic_regression))
conf_ints <- exp(confint(logistic_regression, level = 0.95)) 

# Combine the odds ratios and confidence intervals into a data frame
odds_ratio_table <- data.frame(Odds_Ratio = odds_ratios, 
                                Lower_CI = conf_ints[,1], 
                                Upper_CI = conf_ints[,2],
                                Std_Error = summary(logistic_regression)$coef[,2])

print(odds_ratio_table)
````

````r
# Determine colors for odds ratios
level <- 0.075 # Totalement arbitraire pour l'instant
colors <- ifelse(odds_ratio_table$Lower_CI >= 1 - level & odds_ratio_table$Upper_CI <= 1 + level, "green",
                 ifelse(odds_ratio_table$Odds_Ratio > 1, "red", "blue")) # Peut-être à revoir

# Plot the odds ratios with colors
ggplot(odds_ratio_table, aes(x = rownames(odds_ratio_table), y = Odds_Ratio, fill = colors)) +
  geom_bar(stat = "identity", color = "black") +
  geom_errorbar(aes(ymin = Lower_CI, ymax = Upper_CI), width = 0.2, color = "black") +
  labs(title = "Odds Ratios for Diabetes", x = "Variable", y = "Odds Ratio") +
  theme_classic() +
  theme(axis.text.x = element_text(angle = 45, vjust = 1, hjust = 1))
````

### Support Vector Machine (SVM)

https://www.projectpro.io/recipes/use-svm-classifier-r Lien utile, parle de diabète

Support Vector Machines is a supervised learning algorithm which can work for both classification and regression problems. The main objective of the SVM is to find the optimum hyperplane (i.e. 2D line and 3D plane) that maximises the margin (i.e. twice the distance between the closest data point and hyperplane) between two classes.

It applies a penalty for misclassification and transforms the data to higher dimension if it is non-linearly separable. ​

Major advantages of using SVM are that: ​

- it works well with large number of predictors.
- Works well in-case of non-linear separable data.
- Works well in-case of image classification and does not suffer multicollinearity problem.

This recipe demonstrates the modelling of a SVM Classifier for Binary classification, we use a famous dataset by National institute of Diabetes and Digestive and Kidney Diseases. ​
